#Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?   
- Answer:TextBlob is an open-source Python library built on top of NLTK and Pattern. It provides a simple, user-friendly API for diving into common NLP tasks, acting as a wrapper that hides the complex underlying mechanics.  Simplification of Sentiment Analysis: TextBlob simplifies sentiment analysis by providing a straightforward .sentiment property on any text object. It bypasses the need to manually tokenize, clean text, or train a statistical classifier. It instantly returns two metrics:  Polarity: A float ranging from -1.0 (highly negative) to +1.0 (highly positive).  Subjectivity: A float ranging from 0.0 (very objective/factual) to 1.0 (very subjective/opinion-based).  Simplification of Translation: Behind the scenes, TextBlob interfaces with external translation APIs (such as the Google Translate API). Users can translate text into different languages or detect the source language using a single method call (e.g., .translate(to='es')), completely abstracting away complex machine translation model training and deployment.

#Question 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).  
-   Answer:A Recurrent Neural Network (RNN) is a class of artificial neural networks designed specifically to process sequential data, such as text, time-series, or audio. Unlike traditional feedforward neural networks that assume all inputs are independent of each other, RNNs retain information about past inputs using internal memory loops.  ArchitectureAn RNN can be visualized as a standard feedforward network duplicated over time, or as a single network layer containing a recurrent loop that feeds its output back into itself. At any given time step $t$, the network layer contains:An input vector $x_t$ (e.g., the current word embedding).A hidden state vector $h_t$, which acts as the network's "memory".An output vector $y_t$.FunctioningHidden State Update: At time step $t$, the network receives the current input $x_t$ and the previous hidden state $h_{t-1}$. It combines them using weight matrices ($W_{hh}$ and $W_{xh}$) and applies an activation function (typically $\tanh$):$$h_t = \tanh(W_{hh} h_{t-1} + W_{xh} x_t + b_h)$$Output Generation: The current output $y_t$ is computed using the newly updated hidden state and an output weight matrix ($W_{hy}$):$$y_t = \text{softmax}(W_{hy} h_t + b_y)$$Sequential Propagation: This process repeats step-by-step across the entire sequence length.Because the weights are shared across all time steps, the network can theoretically capture context from many steps back, making it capable of understanding sentence structure. However, in practice, vanilla RNNs suffer from the vanishing gradient problem, which prevents them from remembering long-term dependencies

#Question 5: What is the key difference between LSTM and GRU networks in NLP applications?   
-  Answer:The key difference between Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks lies in their structural complexity, architectural gating mechanisms, and performance efficiency. Both were designed to solve the vanishing gradient problem of vanilla RNNs, but they do so differently:  Gating Structure & Components: An LSTM utilizes three gates (Input gate, Forget gate, Output gate) and manages two distinct states: the Hidden State ($h_t$) for short-term memory and the Cell State ($c_t$) for long-term memory. A GRU simplifies this by combining the cell state and hidden state into a single hidden state ($h_t$) and utilizes only two gates: an Update gate (which controls what to keep and what to throw away) and a Reset gate.Computational Efficiency: Because a GRU has fewer parameters and gates to train, it is computationally faster and requires less memory/training data than an LSTM.NLP Expressiveness: LSTMs can theoretically capture more complex, subtle linguistic nuances and long-term context because of their dedicated cell state. However, on most standard NLP tasks (like text classification or sentiment analysis), GRUs perform comparably well while saving significant computational time.

```python
#Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive(1) or negative (0).
# Dataset texts '''= [
'''“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.'''
#ANSWER
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Dataset Initialization
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

# 2. Preprocessing & Tokenization
vocab_size = 50
max_length = 6
trunc_type = 'post'
padding_type = 'post'

tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
word_index = tokenizer.word_index

# Convert text sentences to sequences of integers
sequences = tokenizer.texts_to_sequences(texts)
# Pad sequences so they all share the same input dimension length
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

# 3. Build the LSTM Model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8, input_length=max_length),
    LSTM(units=8, dropout=0.2),
    Dense(units=1, activation='sigmoid') # Binary classification output
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. Train the Model
print("Training the LSTM model...")
history = model.fit(padded_sequences, labels, epochs=5, verbose=1)

print("\nModel training complete.")
```

```python
#9: Using spaCy, build a simple NLP pipeline that includes tokenization, lemmatization, and entity recognition. Use the following paragraph as your dataset:
'''“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”
Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.'''
#ANSWER
import spacy

# Load the core English spaCy pipeline
nlp = spacy.load("en_core_web_sm")

# Input text dataset
text = (
    "Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the "
    "development of India's atomic energy program. He was the founding director of the Tata "
    "Institute of Fundamental Research (TIFR) and was instrumental in establishing the "
    "Atomic Energy Commission of India."
)

# Process the text using the spaCy pipeline
doc = nlp(text)

# 1. Print Tokens and Lemmatization
print("--- TOKENS AND LEMMAS ---")
print(f"{'Token':<20} | {'Lemma':<20}")
print("-" * 43)
# Showing first 10 for display brevity in the document
for token in list(doc)[:12]:
    print(f"{token.text:<20} | {token.lemma_:<20}")
print("... (truncated for layout) ...\n")

# 2. Print Named Entity Recognition (NER)
print("--- NAMED ENTITIES FOUND ---")
print(f"{'Entity Text':<40} | {'Label/Type':<15}")
print("-" * 60)
for ent in doc.ents:
    print(f"{ent.text:<40} | {ent.label_:<15}")
```

```python
#uestion 10: You are working on a chatbot for a mental health platform. Explain how
'''you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
NLP to understand and respond to user input effectively. Detail your architecture, data
preprocessing pipeline, and any ethical considerations'''
#ANSWER-'''System ArchitectureWe will deploy a Hybrid Intent-Response Architecture:NLP Understanding Layer (spaCy): Used for lightning-fast tokenization, lemmatization, and Named Entity Recognition (to catch names, medications, dates).  Deep Learning Sequence Layer (GRU Network): We opt for a bidirectional GRU network over an LSTM. GRU's lighter profile enables lower latency for real-time chat while matching LSTM text classification accuracy. The GRU will categorize user states into explicit emotional intents (e.g., Anxiety, Depression, Crisis/Self-Harm, General Conversation).  Dialogue Manager: An intent router that decides whether to trigger a safe scripted response, an empathetic dynamic output, or immediate human crisis intervention'''
import spacy
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load spaCy for deep cleanup
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

def preprocess_mental_health_input(user_raw_text, tokenizer, max_len=20):
    """
    Cleans raw user input, lemmatizes words to preserve base sentiment tokens,
    and converts text to padded sequences for the GRU classifier.
    """
    # Remove anomalies, normalize text case
    doc = nlp(user_raw_text.lower().strip())

    # Extract lemmas, removing punctuation and stop words that don't add emotional context
    cleaned_tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    cleaned_string = " ".join(cleaned_tokens)

    # Transform to numerical sequences and pad
    sequence = tokenizer.texts_to_sequences([cleaned_string])
    padded_seq = pad_sequences(sequence, maxlen=max_len, padding='post')

    return padded_seq

print("Preprocessing pipeline initialized successfully.")
```
